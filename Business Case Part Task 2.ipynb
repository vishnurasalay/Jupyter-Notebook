Import Pandas Library

```python
import pandas as pd
```

Read Excel File From Source

```python
df = pd.read_excel(r"C:\Users\rvishnuv\Downloads\Python_test\Python_test\interval_data.xlsx")
```

```python
df
```

Create Data Frame With Columns MPAN, Date, Hour, and Value

```python
melted_df = pd.melt(df, id_vars=['MPAN', 'Date'], var_name='Hour', value_name='Value')
```

```python
print(melted_df)
```

```python
melted_df.sort_values(by='Value', ascending=True)
```

```python
melted_df['Date'] = pd.to_datetime(melted_df['Date'])
```

```python
melted_df['week'] = melted_df['Date'].dt.week
```

```python
print(melted_df)
```

```python
melted_df['week'].unique()
```

```python
result = melted_df.groupby(['MPAN', 'week'])['Value'].agg(['mean', 'max', 'min']).reset_index()
```

```python
result
```

```python
output_folder = r'C:\Users\rvishnuv\Downloads\Python_test'

# Group by 'MPAN'
grouped_data = result.groupby('MPAN')

# Iterate through groups and save to separate files
for mpan, group in grouped_data:
    filename = f"{output_folder}result_{mpan}.csv"
    group.to_csv(filename, index=False)

print("Files saved successfully.")
```
